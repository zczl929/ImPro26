## Task 1: Compute Histogram
Histograms can be rather important when it comes to adjusting the brightness of an image.
Your first task is to compute a histogram of an image of your choice and display this as a plot.

*Note: Do not use any libraries, write your own function*

```python
def compute_hist(img):
    '''
    Impelemnt 
    :param img: gray scaled image with size HxW
    :return: 
    '''
    histogram = np.zeros(256) # create a histogram with 256 bins for each pixel value from 0 to 255
    print(histogram)
    
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            histogram[img[i][j]] += 1
    
    return histogram 
```

## Task 2: Histogram Equalisation
Histogram Equalisation is often used to adjust the global contrast of an image.
It is most useful when an images background is either too dark or too light and can bring out more detail in under exposed images.

This can be done by finding a mapping function $F: [0, 255] \rightarrow [0, 255]$:

$$
    F(v) = \frac{cdf(v) - cdf_{min}}{(M \times N) - cdf_{min}}x(L - 1)
$$

Where $v$ is the current value to be adjusted, $(M \times N)$ is the size of the image and $L$ is the number of intensity levels i.e. 256. The cdf is the cumulative distribution of the histogram.

Your task would be to calculate the new mappings, apply this to the image and plot the new equalised histogram. As shown below:

```python
#     
#    (contrast)    ,       .
def hist_equalisation(img, img_hist):
    '''
    Impelemnt 
    :param img: gray scaled image with size HxW
    :param img_hist: [256,]
    :param L:
    :return: image segmentation of size HxW
    '''
    # Number of intensity levels
    L = 256
    
    # Total number of pixels (M x N)
    M, N = img.shape
    total_pixels = M * N
    
    # Step 1: Calculate the cumulative distribution function (CDF)
    cdf = [0] * L  
    cdf[0] = img_hist[0] 
    for i in range(1, L):
        cdf[i] = cdf[i-1] + img_hist[i]
        
    # Step 2: Normalize the CDF
    cdf_min = min(cdf) # get the minimum value of the CDF
    equalization_map = [0] * L  # create a mapping array to map the old pixel values to new pixel values
    for v in range(L):
        equalization_map[v] = round(((cdf[v] - cdf_min) / (total_pixels - cdf_min)) * (L - 1))
    
    # Step 3: Apply the equalization map to the image
    equalized_img = img.copy()
    for i in range(M):
        for j in range(N):
            equalized_img[i, j] = equalization_map[img[i, j]] 
    
    # Step 4: Compute the histogram of the equalized image
    equalized_hist = [0] * L 
    for row in equalized_img:
        for pixel in row:
            equalized_hist[pixel] += 1 
    
    # Plot the new histogram
    plt.figure(figsize=(8, 6))
    plt.bar(range(256), equalized_hist, color='gray')
    plt.title('Equalized Histogram')
    plt.xlabel('Pixel Intensity (0-255)')
    plt.ylabel('Frequency')
    plt.show()
    
    return equalized_img

# Example usage:
# Assume img is a 2D numpy array representing a grayscale image
# and img_hist is the histogram of that image
# equalized_image = hist_equalisation(img, img_hist)
```

```python
eqImg = hist_equalisation(gray_img, img_hist)
eqImg_hist = compute_hist(eqImg.astype(int))

# Plot Original and Equalised image
plt.figure(figsize=(15,10))

plt.subplot(221)
plt.imshow(gray_img, 'gray')
plt.grid(False)
plt.title('Original Image')

plt.subplot(222)
plt.imshow(eqImg, 'gray')
plt.grid(False)
plt.title('Equalized Image')

plt.subplot(212)
plt.plot(eqImg_hist, color='cyan', label='equalised')
plt.plot(img_hist, label='original')
plt.legend()

plt.show()
```

## Task 3: Histogram Matching

Your next task is to be able to adjust an image based on the histogram of another.
This is done using histogram matching and can be helpful if you would like to adjust a series of images with different global illuminations. Using histogram matching you could adjust this images brightness (and contrast) to match the other better illuminated images.

To perform histogram matching we need to first calculate the normalised CDF of the **source image** and the **target image**: `cdfSource` and `cdfTarget`.

The normalised CDF for each image is obtained by dividing the value in each bin of the original CDF with the maximum value of the original CDF (last bin of the original CDF), so we get a value between 0 and 1 in each bin of the normalised CDF.

We then need to find a mapping function `F` that maps `cdfSource` to `cdfTarget`. This mapping function can then be applied to the input image. The mapping function needed is quite simple and can be obtained by comparing the values between cdfSource and cdfTarget. One way to do this, is to find the first `cdfSource(valueS)` that is greater than `cdfTarget(valueT)`, and that valueT will be the mapping `F(valueS)`. The figure below is a demonstration on how this mapping function works:

<img src="../images/histogram-cdf.png" alt="drawing" width="700"/>

```python
import numpy as np 
import matplotlib.pyplot as plt 

def compute_cdf(hist):
    ''' 
    CDF  
    :param hist: histogram with size 256 (  )
    :return: cdf with size 256 (   )
    '''
    cdf = np.cumsum(hist) # compute CDF
    cdf_normalized = cdf / cdf[-1] # normalize CDF
    return cdf_normalized 
```

```python
# histogram matching function 
def hist_match(source_img, target_img):
    '''
    Impelemnt 
    :param source_img: gray scaled image with size HxW
    :param target_img: 
    :return: image 
    '''
    source_hist = [0] * 256
    target_hist = [0] * 256 
    
    for row in source_img: 
        for pixel in row: 
            source_hist[pixel] += 1 # source image histogram
    
    for row in target_img: 
        for pixel in row: 
            target_hist[pixel] += 1 # target image histogram
    
    # compute CDFs 
    source_cdf = compute_cdf(source_hist)
    target_cdf = compute_cdf(target_hist)
    
    # create a mapping table
    mapping = np.zeros(256, dtype=int) 
    valueT = 0 
    
    for valueS in range(256):
        while target_cdf[valueT] < source_cdf[valueS] and valueT < 255:
            valueT += 1
        mapping[valueS] = valueT
    
    # apply the mapping table to the source image
    matched_img = np.zeros_like(source_img)
    
    for i in range(source_img.shape[0]):
        for j in range(source_img.shape[1]):
            matched_img[i, j] = mapping[source_img[i, j]] # map the pixel values
        
    return matched_img, source_hist, target_hist
```

```python
src_img = cv2.imread('../images/meerkat.jpg')
src_img = cv2.cvtColor(src_img, cv2.COLOR_BGR2GRAY)
tgt_img = gray_img
tgt_hist = img_hist

matched_img, src_hist, tag_hist = hist_match(src_img, tgt_img)
matched_hist = compute_hist(matched_img.astype(int))


# Plot Input, Output and Matched image
plt.figure(figsize=(15,10))

plt.subplot(231)
plt.imshow(src_img, 'gray')
plt.grid(False)
plt.title('Input Image')

plt.subplot(232)
plt.imshow(gray_img, 'gray')
plt.grid(False)
plt.title('Target Image')

plt.subplot(233)
plt.imshow(matched_img, 'gray')
plt.grid(False)
plt.title('Matched Image')

plt.subplot(212)
plt.plot(tgt_hist, color='cyan', label='target')
plt.plot(src_hist, label='input')
plt.plot(matched_hist, label='matched')
plt.legend()

plt.show()
```
